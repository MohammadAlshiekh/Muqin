### Extracting the dialog from each character

```python
file_path = './مجنون ليلى_chapters/chapter_1.txt'
```

```python
def extract_dialogues_corrected(file_path):
    dialogues_by_character = {}
    current_character = None
    collecting = False
    dialogue_lines = []

    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            line = line.strip()
            # Check if we are at a character name line
            if not collecting and not line.endswith(':') and line.isalpha():
                current_character = line
            # Check if the line is just a colon, indicating the start of dialogue
            elif line == ':':
                collecting = True
                dialogue_lines = []  # Reset for new dialogue
            # If we're collecting dialogue and hit an empty line or new character, stop collecting
            elif collecting and (not line or (line.isalpha() and not line.endswith(':'))):
                # Save dialogue if we have collected some lines
                if dialogue_lines:
                    dialogues_by_character[current_character] = dialogues_by_character.get(current_character, []) + [" ".join(dialogue_lines)]
                collecting = False
            elif collecting:
                # Add dialogue line
                dialogue_lines.append(line)

    return dialogues_by_character

dialogues = extract_dialogues_corrected(file_path)

# Print the dialogues
for character, dialogues_list in dialogues.items():
    print(f"{character}:")
    for dialogue in dialogues_list:
        print(f"  '{dialogue}'")
    print("\n---\n")
```
